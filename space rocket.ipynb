```python
from stable_baselines3 import PPO,A2C	
```

```python
import gym
gym.logger.MIN_LEVEL = 1
```

```python
env = gym.make( "LunarLander-v2")
```

```python
for i in range(5):
    done=False
    state =env.reset()
    while not done:
        
        state,reward,done,info=env.step(env.action_space.sample())
        # print(reward,done,info)
        env.render()
env.close()
```

```python
model = PPO("MlpPolicy", env, verbose=1)
model = A2C("MlpPolicy", env, verbose=1)
```

```python
model.learn(total_timesteps=25_000*2)
```

```python
vec_env = model.get_env()
obs = vec_env.reset()
for i in range(20):
    done=False
    obs = vec_env.reset()
    reward_sum=0
    while not done:
        action, _states = model.predict(obs, deterministic=True)
        obs, reward, done, info = vec_env.step(action)
        vec_env.render()
        reward_sum+=reward
    print(reward_sum)
    # VecEnv resets automatically
    # if done:
    #   obs = env.reset()

env.close()
```
